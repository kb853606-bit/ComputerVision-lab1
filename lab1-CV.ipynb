```python
# Step 1: Import required libraries
from skimage import io, color, filters, feature
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
# Step 2: Upload and read the image
print("Please upload your image (object, face, or building):")
uploaded = files.upload()

# Read the uploaded image
image_name = list(uploaded.keys())[0]
img = io.imread(image_name)
```

```python

# Step 3: Convert image to grayscale
gray = color.rgb2gray(img)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python

# Step 4: Apply Gaussian filter with different sigma values
sigma_values = [1, 2, 4]
blurred_images = []

plt.figure(figsize=(15,5))
for i, s in enumerate(sigma_values):
    blurred = filters.gaussian(gray, sigma=s)
    blurred_images.append(blurred)
    plt.subplot(1,3,i+1)
    plt.imshow(blurred, cmap='gray')
    plt.title(f"Gaussian Filter (sigma={s})")
    plt.axis('off')
plt.show()
```

```python
# Step 5: Perform Canny edge detection

plt.figure(figsize=(15,5))
for i, s in enumerate(sigma_values):
    edges = feature.canny(blurred_images[i], sigma=1)
    plt.subplot(1,3,i+1)
    plt.imshow(edges, cmap='gray')
    plt.title(f"Canny after sigma={s}")
    plt.axis('off')
plt.show()
```

## Observation
When the sigma value is small (1), the image keeps more details, and more edges are visible.  
When sigma increases (like 4), the image becomes smoother, and fewer edges are detected because small details are blurred.
